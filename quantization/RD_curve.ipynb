```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def CodeRate(delta, f, x):
  entropy_x = np.trapezoid(f(x), x)
  return entropy_x - np.log2(delta)

def Distortion(delta, f):
  K = range(-100, 100)
  U = np.linspace(- delta / 2, delta / 2, 1000)
  S = np.sum([[u**2 * f(u + (k + 1/2) * delta) for u in U] for k in K], axis=0)

  D = np.trapezoid(S, U)
  return D

def f(x):
  return np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)
```

```python
N = 10000
delta = np.linspace(0.1, 1, 10)
x = np.linspace(-5, 5, N)

rate = []
distortion = []

for d in delta:
  rate.append(CodeRate(d, f, x))
  distortion.append(Distortion(d, f))
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(distortion, rate, marker='o')
plt.title('Code Rate vs distortion')
plt.xlabel('Distortion')
plt.ylabel('Code Rate (bits)')
plt.grid()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(delta, rate, marker='o')
plt.title('Code Rate vs delta')
plt.xlabel('Delta')
plt.ylabel('Code Rate (bits)')
plt.grid()
```
